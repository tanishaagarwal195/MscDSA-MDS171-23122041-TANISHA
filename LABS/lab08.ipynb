# Lab 08
Matrix Multiplication

Jeeva, a mathematical expert, is looking for an expert in programming who can assist him in creating a Python program for Matrix Multiplication.

His program has specific requirements.

The matrices have to be saved inside a dictionary.
The keys allowed in the dictionary are [matrix 1, matrix 2, result matrix]
The values for the elements in the rows of the matrix have to be saved in the form of a list of lists [[row1][row2]].
You must have functions for collecting the matrix and store in the dictionary.
Another function to Process the matrix and store it in the dictionary. 
Another function to Display the result in a file, properly formatted.
Menu option to show enter matrix 1, matrix 2, calculate, export to file. If the matrix has been entered, then it should take confirmation from the user before taking the input for matrix 1 or 2 again

```python
# Initialize empty dictionaries to store matrices
matrix_dict = {
    "matrix 1": None,
    "matrix 2": None,
    "result matrix": None
}

# Function to collect a matrix from the user and store it in the dictionary
def collect_matrix(matrix_key):
    rows = int(input(f"Enter the number of rows for {matrix_key}: "))
    cols = int(input(f"Enter the number of columns for {matrix_key}: "))
    matrix = []

    for i in range(rows):
        row = []
        for j in range(cols):
            value = float(input(f"Enter the value at row {i+1}, column {j+1}: "))
            row.append(value)
        matrix.append(row)

    matrix_dict[matrix_key] = matrix

# Function to perform matrix multiplication and store the result in the dictionary
def multiply_matrices():
    if matrix_dict["matrix 1"] is None or matrix_dict["matrix 2"] is None:
        print("Please enter both Matrix 1 and Matrix 2 first.")
        return  

    matrix1 = matrix_dict["matrix 1"]
    matrix2 = matrix_dict["matrix 2"]

    if len(matrix1[0]) != len(matrix2):
        print("Matrix dimensions are not suitable for multiplication.")
        return

    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            dot_product = sum(matrix1[i][k] * matrix2[k][j] for k in range(len(matrix2)))
            row.append(dot_product)
        result.append(row)

    matrix_dict["result matrix"] = result
    print("Matrix multiplication complete.")

# Function to display the result matrix in a properly formatted file
def export_to_file():
    result_matrix = matrix_dict["result matrix"]
    if result_matrix is None:
        print("Result matrix is empty. Please calculate it first.")
        return

    filename = input("Enter the filename to save the result: ")
    with open(filename, "w") as file:
        for row in result_matrix:
            file.write("\t".join(map(str, row)) + "\n")
    print(f"Result matrix saved in {filename}.")

# Main menu
while True:
    print("\nMenu Options:")
    print("1. Enter Matrix 1")
    print("2. Enter Matrix 2")
    print("3. Calculate")
    print("4. Export to File")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        collect_matrix("matrix 1")
    elif choice == "2":
        collect_matrix("matrix 2")
    elif choice == "3":
        multiply_matrices()
    elif choice == "4":
        export_to_file()
    elif choice == "5":
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```
